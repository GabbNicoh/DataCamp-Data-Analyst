# Pokemon Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv("C:\\Users\\sawcon\\_datacamp\\datasets\\pokemon-data.csv", usecols=['Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Legendary'])
pokemon.head()
```

```python
pokemon[['Type 1', 'Type 2']].value_counts()
```

What is the percentage Flying Legendary pokemons?

```python
legendary = pokemon[pokemon['Legendary'] == True]

legendary['Type 1'].value_counts(normalize=True).plot(kind='barh', alpha=0.5, color='Orange')
legendary['Type 2'].value_counts(normalize=True).plot(kind='barh', alpha=0.5)

plt.show()
```
